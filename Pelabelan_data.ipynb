```python
import pandas as pd
import numpy as np

def load_data():
  data = pd.read_csv('hasil_preprocessing_data_kpu.csv')
  return data

data = load_data()
data.head()
```

```python
df = pd.DataFrame(data['stemming_data'])
df.head()
```

```python
!pip install VaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

melakukan analisis sentimen pada teks yang terdapat dalam kolom 'stemming_data' menggunakan suatu alat analisis sentimen (diasumsikan sebagai analyser). Skor sentimen kemudian disimpan dalam kolom baru 'Compound_Score'


```python
scores = [analyser.polarity_scores(x) for x in df['stemming_data']]
df['Compound_Score'] = [x['compound'] for x in scores]

df.head(100)
```

```python
df.loc[df['Compound_Score'] < 0, 'Sentiments'] = 'Negatif'
df.loc[df['Compound_Score'] == 0, 'Sentiments'] = 'Netral'
df.loc[df['Compound_Score'] > 0, 'Sentiments'] = 'Positif'

df.head(100)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download resource vader_lexicon
nltk.download('vader_lexicon')


# Fungsi untuk menentukan sentimen positif
def get_sentiment(stemming_data):
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(stemming_data)['compound']
    return sentiment_score > 0

# Mengambil hanya teks dengan sentimen positif
positive_data = df[df['stemming_data'].apply(get_sentiment)]

# Menggabungkan semua teks menjadi satu string
all_text = ' '.join(positive_data['stemming_data'])

# Membuat WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Menampilkan WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Fungsi untuk menentukan sentimen netral
def get_sentiment(stemming_data):
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(stemming_data)['compound']
    return sentiment_score == 0

# Mengambil hanya teks dengan sentimen netral
netral_data = df[df['stemming_data'].apply(get_sentiment)]

# Menggabungkan semua teks menjadi satu string
all_text = ' '.join(netral_data['stemming_data'])

# Membuat WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Menampilkan WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
